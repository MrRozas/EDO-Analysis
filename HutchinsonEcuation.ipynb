```python
# MODEL, WITH UNKNOWN PARAMETERS, http://zulko.github.io/blog/2013/10/22/delay-differential-equations-in-python/

#Instalar lo necesario por 1ra vez
#!pip install ddeint
#!pip install matplotlib

from pylab import *
from ddeint import ddeint

model = lambda Y,t, k,d,r :  (r*Y(t))*(1-(Y(t-d)/k)) #Ecuación de Hutchinson, Y(t) es la ecuación, t es el tiempo, k es la capacidad de carga (K), d es el retardo, y r es la razón de crecimiento intrínseco de la especie 
g = lambda t:3 # Valor inicial del PVI

tt = linspace(0,1000,10000) #El espacio donde se va a graficar, donde parte en eje x, hasta donde llega en eje x, el último no es necesario ser modificado  
yy = ddeint(model, g, tt, fargs=(100, 20, 0.1)) # Asignar valores :K=100 capacidad de carga, d=20 retardo, r=0.1 razón de crecimiento intrínseco

plot(tt,yy,lw=2) #Graficamos
```
